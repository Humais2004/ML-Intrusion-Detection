Data Preprocessing


1. Dropping Unnecessary Columns

```python
def dropInfinateNul(df):
  print(df.shape)

  #replace infinity values as null
  df = df.replace(["Infinity", "infinity"],np.inf)
  df = df.replace([np.inf, -np.inf], np.nan)

  # drop all nan (null values)
  df.dropna(inplace=True)
  print(df.shape)
  return df
```

```python
df1 = dropInfinateNul(df1)
df2 = dropInfinateNul(df2)
df3 = dropInfinateNul(df3)
df5 = dropInfinateNul(df5)
df6 = dropInfinateNul(df6)
```

```python
def dropUnnecessaryColumns(df):
  df.drop(['Timestamp'], axis=1, inplace=True)
  return df
```

```python
df1 = dropUnnecessaryColumns(df1)
df2 = dropUnnecessaryColumns(df2)
df3 = dropUnnecessaryColumns(df3)
df5 = dropUnnecessaryColumns(df5)
df6 = dropUnnecessaryColumns(df6)
```
